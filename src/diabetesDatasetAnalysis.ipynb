```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('dia.csv')
data.head()
```

```python
sns.histplot(data=data, x='Glucose', hue='Outcome', kde=False,palette=['blue', 'green'])
plt.title('Distribution of Glucose Levels by Outcome')
plt.xlabel('Glucose Level')
plt.ylabel('Count')
plt.legend(title='Outcome', loc='upper right',labels=['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
sns.scatterplot(data=data, x='Age', y='BMI', hue='Outcome')
plt.title('BMI vs. Age')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='Outcome', loc='upper right')
plt.show()
```

```python
sns.violinplot(data, x='Outcome', y='BloodPressure')
plt.title('Blood Pressure by Outcome')
plt.xlabel('Outcome')
plt.ylabel('Blood Pressure')
plt.xticks(ticks=[0, 1], labels=['Non-Diabetic', 'Diabetic'])
plt.show()
```

```python
sns.scatterplot(x='Pregnancies', y='DiabetesPedigreeFunction', data=data, alpha=0.5)
plt.title('Scatter Plot between Pregnancies and DiabetesPedigreeFunction (Seaborn)')
plt.show()
```
