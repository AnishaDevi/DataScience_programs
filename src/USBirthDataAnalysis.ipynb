```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

births=pd.read_csv('births.csv')
print(births.head())
print(births.dtypes)
```

```python
births_data=births.groupby(['year','gender'])
print(births_data['births'].sum())
```

```python
from numpy import matrix
births['decade']=10*(births['year']//10)
births.pivot_table('births',index='decade',columns='gender',aggfunc='sum')
```

```python
births['date']=pd.to_datetime(births[['year','month','day']],errors='coerce')
births['date']=pd.to_datetime(births['date'])
births['day_of_week']=births['date'].dt.day_name()
births['decade']=(births['year']//10)*10

avg_daily_by_day_and_decade=births.groupby(['day_of_week','decade'])['births'].mean()
avg_daily_by_day_and_decade
```
